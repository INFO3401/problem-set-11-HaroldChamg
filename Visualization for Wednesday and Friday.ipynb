```python
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
from vega_datasets import data
cars = data.cars()
```

```python
alt.Chart(cars).mark_point().encode(
    x= "Horsepower",
    y= "Miles_per_Gallon",
    color = "Origin"
)
```

```python
alt.Chart(cars).mark_point().encode(
    x= "Horsepower",
    y= "Miles_per_Gallon",
    color = "Origin"
).interactive()
```

```python
alt.Chart(cars).mark_bar().encode(
    x= "Year",
    y= "average(Cylinders)"
)
```

```python
alt.Chart(cars).mark_bar().encode(
    alt.X ("Horsepower",bin = True),
    y= "count()"
)
```

```python
alt.Chart(cars).mark_point().encode(
    x= "Horsepower",
    y= "Acceleration",
    color = "Origin",
    size ="Year"
).interactive()
```

```python
count = pd.read_csv('rawCounts.csv')
count
```

```python
df_groupby = count.groupby(['Author', 'Word']).sum()
df_groupby
```

```python
sort = df_groupby.reset_index().sort_values(['Author', 'Count'], ascending = False).set_index(['Author', 'Word'])
sort.head(10)
```

```python
most_count = sort.groupby('Author').head(100).reset_index()
most_count
```

```python
alt.Chart(most_count).mark_bar().encode(
    x='Author',
    y='Count',
).interactive()
```
